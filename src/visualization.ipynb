Introduction
This notebook is a practical introduction to using Matplotlib and Seaborn for data visualization in Python.
We will cover:
 - Basic plotting with Matplotlib
 - Customizing plots
 - Creating more advanced visualizations with Seaborn

```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python


# Generate Sample Data
data = pd.DataFrame({
    'x': np.linspace(0, 10, 100),
    'y': np.sin(np.linspace(0, 10, 100)) + np.random.normal(0, 0.2, 100),
    'category': np.random.choice(['A', 'B'], size=100)
})

# Matplotlib Basics
```

```python
data
```

```python

# ### Line Plot
plt.figure(figsize=(8, 5))
plt.plot(data['x'], data['y'], label='Line', color='blue')
plt.plot(data['x'] + 1, data['y'] + 1, label='x + 1', color='red')
plt.title('Line Plot with Matplotlib')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(False)
plt.show()
```

```python

# ### Scatter Plot
plt.figure(figsize=(8, 5))
plt.scatter(data['x'], data['y'], c='green', alpha=0.6, s=50)
plt.title('Scatter Plot with Matplotlib')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()

```

```python

# ## Seaborn Basics

# ### Scatter Plot with Hue
sns.scatterplot(data=data, x='x', y='y', hue='category', s=100)
plt.title('Seaborn Scatter Plot with Hue')
plt.show()

```

```python


# ### Line Plot
sns.lineplot(data=data, x='x', y='y', ci='sd', lw=10)
plt.title('Seaborn Line Plot with Confidence Interval')
plt.show()


```

```python

# ### Box Plot
sns.boxplot(x='category', y='y', data=data)
plt.title('Seaborn Box Plot by Category')
plt.show()

```

```python

# ## Customizing Seaborn Plots
plt.figure(figsize=(10, 5))
sns.scatterplot(data=data, x='x', y='y', hue='category', style='category', palette='deep')
plt.title('Customized Seaborn Plot')
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.legend(title='Category')
plt.show()

# ## Conclusion
# - Matplotlib gives fine-grained control over plots
# - Seaborn makes complex plots easier and prettier
# - Both libraries are essential for data visualization in Python

# Try customizing the plots or using your own dataset to explore more!
```

```python

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10, 4))

# First subplot (1 row, 2 columns, 1st plot)
plt.subplot(1, 3, 1)
plt.plot(x, y1, label='sin(x)')
plt.title('Sin Plot')
plt.legend()

# Second subplot (1 row, 2 columns, 2nd plot)
plt.subplot(1, 3, 2)
plt.plot(x, y2, label='cos(x)', color='orange')
plt.title('Cos Plot')
plt.legend()

# Second subplot (1 row, 2 columns, 2nd plot)
plt.subplot(1, 3, 3)
plt.plot(x, y2, label='cos(x)', color='purple')
plt.title('Cos Plot')
plt.legend()


plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 4))  # 1 row, 2 columns

# First plot
axes[0,0].plot(x, y1, label='sin(x)')
axes[0,0].set_title('Sin Plot')
axes[0,0].legend()

# Second plot
axes[0,1].plot(x, y2, label='cos(x)', color='orange')
axes[0,1].set_title('Cos Plot')
axes[0,1].legend()

# Third plot
axes[1,0].plot(x, y2, label='cos(x)', color='red', lw=70)
axes[1,0].set_title('Cos Plot')
axes[1,0].legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)               # First Y
y2 = 100 * np.exp(-0.3 * x)  # Second Y (different scale)

fig, ax1 = plt.subplots()

# First Y-axis
ax1.plot(x, y1, color='blue', label='sin(x)')
ax1.set_ylabel('sin(x)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Second Y-axis
ax2 = ax1.twinx()  # Creates a second Y-axis sharing the same x-axis
ax2.plot(x, y2, color='red', label='exp decay')
ax2.set_ylabel('exp(-0.3x)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.title('Dual Y-Axis Plot')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D  # Needed for 3D plotting

# Generate some example data
np.random.seed(0)
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

# Create a 3D plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the points
scatter = ax.scatter(x, y, z, c=z, cmap='viridis', s=70, alpha=0.09)

# Add labels
ax.set_title('3D Scatter Plot')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Add color bar
fig.colorbar(scatter, ax=ax, label='Z value')

plt.show()
```
